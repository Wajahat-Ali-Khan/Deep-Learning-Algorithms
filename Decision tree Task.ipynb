```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
# Load the iris dataset
iris = pd.read_csv("Iris.csv")
```

```python
iris.tail()
```

```python
x_train, x_val, y_train, y_val = train_test_split(iris.iloc[:, :-1].values,  iris.iloc[:, -1].values, test_size=0.2, random_state = 0)
```

```python
x_train.shape
```

```python
dtree=DecisionTreeClassifier()
ftree=dtree.fit(x_train,y_train)
```

```python
type(dtree)
```

```python
from sklearn import tree
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(ftree, 
                   feature_names=iris.iloc[:, :-1].values,  
                   class_names=iris.iloc[:, -1].values,
                   filled=True)
```

```python
y_pred = ftree.predict(x_val)#Accuracy
from sklearn import metrics
print('Accuracy Score:', metrics.accuracy_score(y_val,y_pred))
```
